# DECISION TREE AND NAIVE BAYES CLASSIFICATION

```python
import pandas as pd 
import numpy as np 
from sklearn.tree import DecisionTreeClassifier 
#to draw the graph
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image
import pydotplus 
```

### Downloading dataset from the given URL

```python
data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names=['sepal_l','sepal_w','petal_l','petal_w','class']
iris_df = pd.read_csv(data_url, names = names)
iris_df.head(100)
```

```python
iris_df['class'].unique()
```

### Replacing the string values of the "class" column into numerics

```python
classes = {
    'Iris-setosa': 0,
    'Iris-versicolor' : 1,
    'Iris-virginica' : 2
    }
iris_df = iris_df.replace({'class': classes})
iris_df.head(100)
```

### Splitting the dependent and independent variables

```python
X = iris_df.iloc[:,:4]
Y = iris_df.iloc[:,4]
#converting it into an array
X = np.array (X)
Y = np.array (Y)
```

```python
X
```

```python
Y
```

```python
iris_df.shape
```

## VISUALIZING DECISION TREE

### Keeping the maximum depth of the decision tree as 4

```python
# 4 levels of splitting will occur
tree = DecisionTreeClassifier(criterion='entropy', max_depth = 4)
tree.fit(X, Y)
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data, filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png(), width=750, height=10)
```

### Using 50 samples to split the decision tree

```python
# minimum 50 samples are required to split the tree
tree = DecisionTreeClassifier(criterion='entropy', min_samples_leaf = 50)
tree.fit(X, Y)
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data, filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png(), width=400, height=10)
```

### Using the minimum number of samples as 10 to split the tree 

```python
# how many samples tree need to have to split = 10
tree = DecisionTreeClassifier(criterion='entropy',min_samples_split = 10)
tree.fit(X, Y)
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data, filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png(), width=700, height=10)
```

### Using entropy as the criterion

```python
#using entropy as the criterion
tree = DecisionTreeClassifier(criterion='entropy')
tree.fit(X, Y)
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data, filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png(), width=700, height=10)
```

### Using Gini Index as the criterion

```python
# using gini index as criterion
tree = DecisionTreeClassifier(criterion='gini')
tree.fit(X, Y)
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data, filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png(), width=700, height=10)
```

### Predicting values using the decision tree

```python
predict = tree.predict(X)
predict
```

```python
from sklearn import metrics
print('Accuracy:', metrics.accuracy_score(Y, predict))
```

## NAIVE BAYES CLASSIFICATION

### Creating the object for Naive Bayes classification

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

### Performing train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2) # 20% test size
nb.fit(X_train,Y_train)
```

### Predicting values using the trained model

```python
Y_pred=nb.predict(X_test)
Y_pred
```

```python
Y_test
```

### Accuracy score

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred)
```
