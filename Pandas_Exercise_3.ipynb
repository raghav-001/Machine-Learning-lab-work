Building DataFrames with Pandas
From Python Dictionaries

```python
list1 = ['S1','S2','S3','S4']
English = [89,90,78,98]
Maths = [67,90,100,100]
Science = [78,89,90,89]

titles = ['ID', 'English','Maths','Science']
values = [list1,English,Maths,Science]
result = list(zip(titles,values))
result
```

```python
result1 = dict(result)
df1 = pd.DataFrame(result1)
print(df1)
```
